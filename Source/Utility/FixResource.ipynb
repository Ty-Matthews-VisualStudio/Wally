```python
# This code fixes duplicate ID assignments in resource.h by regenerating the list and recounting
import os

def printjust(dest, s, id):
    dest.write('#define ' + s.ljust(31) + ' ' + str(id) + '\n')    

source = open(r'..\Wally\resource.h', 'r')
dest = open(r'..\Wally\resource_fixed.h','w')
Lines = source.readlines()
sub100 = 10
resourceid = 100
controlid = 1000
commandid = 2000
customid_1 = 11000
customid_2 = 32000
customid_3 = 61000

for row in Lines:
    if row.lstrip()[:len('#define')].lower() != '#define':
        dest.write(row)
    else:
        true_split = list()
        split = row.strip().split(' ')
        for s in split:
            if len(s) > 0 and s.lower() != '#define':
                true_split.append(s)
                
        if len(true_split) == 2:
            id = int(true_split[1])
            if id > 0 and id < 100:
                printjust(dest, true_split[0], sub100)
                sub100 = sub100 + 1
                
            if id >= 100 and id < 1000:
                printjust(dest, true_split[0], resourceid)
                resourceid = resourceid + 1
                
            if id >= 1000 and id < 2000:
                printjust(dest, true_split[0], controlid)
                controlid = controlid + 1
                
            if id >= 2000 and id < 11000:
                printjust(dest, true_split[0], commandid)
                commandid = commandid + 1
                
            if id >= 11000 and id < 32000:
                printjust(dest, true_split[0], customid_1)
                customid_1 = customid_1 + 1
                
            if id >= 32000 and id < 61000:
                printjust(dest, true_split[0], customid_2)
                customid_2 = customid_2 + 1
                
            if id >= 61000:
                printjust(dest, true_split[0], customid_3)
                customid_3 = customid_3 + 1
                
    
dest.close()
source.close()
```
